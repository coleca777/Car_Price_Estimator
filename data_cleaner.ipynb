```python
import pandas as pd
import warnings

pd.set_option('display.max_columns', None)
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('cleaned_car_data_v2.csv')
```

```python
cols_of_interest = ['price','body_type','exterior_color', 'has_accidents', 'listed_date', 'make_name', 'mileage', 'model_name', 'owner_count', 'year', 'dealer_zip', 
                    'major_options','trim_name','wheel_system']
df1 = df[cols_of_interest]
```

```python
df1['has_accidents'] = df1['has_accidents'].fillna('False')
df1['owner_count'] = df1['owner_count'].fillna('1')
df1
```

```python
for a in cols_of_interest:
    df1[a] = df1[a].astype(str)
    #df1 = df1.loc[(df1[a] != 'nan')]
df1
```

```python


def clen_cols(x):
    if "nan" in x or 'None' in x:
        return False
    return True

df1['temp'] = df1.values.tolist()
df1['temp'] = df1['temp'].apply(clen_cols)

df1
```

```python
df1 = df1.loc[df1['temp'] == True]
df1 = df1.drop(columns=['temp'])
#df1.to_csv('cleaned_car_data.csv', sep = "|", index = False)

df1
```
